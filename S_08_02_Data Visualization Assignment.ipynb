# S_08_02_Data Visualization Assignment

In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

**Task 1** : Transform iris data 4-dimensions into 3-Dimensions  
**Task 2** : Plot a 3D-Chart with transformed dimensions   
**Task 3** : Colour each data point with specific class  

##### Importing required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
%matplotlib inline

from sklearn.decomposition import PCA

from mpl_toolkits.mplot3d import Axes3D
```

##### Loading IRIS data

```python
iris_dataset = datasets.load_iris()

# Separating out the features
X = iris_dataset.data

# Separating out the target
Y = iris_dataset.target
```

**Task 1** : Transform iris data 4-dimensions into 3-Dimensions 
##### Using PCA
=====================================================


```python
# Reducing the dimensions for 3D
pca = PCA(n_components=3)
X = pca.fit_transform(X)
```

**Task 2** : Plot a 3D-Chart with transformed dimensions  
**Task 3** : Colour each data point with specific class

=====================================================


```python
#Visualising the 3D chart
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=Y, cmap=plt.cm.rainbow_r, edgecolor='k')
plt.show()
```
