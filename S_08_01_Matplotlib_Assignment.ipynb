**Curve fitting : Temperature as a function of month of the year**

==============================================================================================================

We have the min and max temperatures in a city In India for each months of the year.   
We would like to find a function to describe this and show it graphically, the dataset given below.

##### Importing required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from scipy import optimize
```

##### Input
We have the min and max temperatures in a city In India for each months of the year.

```python
temp_max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
temp_min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
```

##### Creating months in numbers

```python
months = np.arange(12)
```

##### Plotting the points

```python
plt.plot(months, temp_max, 'ro')
plt.plot(months, temp_min, 'bo')
plt.xlabel('Months')
plt.ylabel('Temperature ($^\circ$C)')
```

##### Task 1 : Fitting it to a periodic function

```python
def yearly_temps(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, months, temp_max, [20, 10, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months, temp_min, [-40, 20, 0])
```

##### Task 2 : Plotting the fit

```python
days = np.linspace(0, 12, num=365)

plt.figure()
plt.plot(months, temp_max, 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'r-')
plt.plot(months, temp_min, 'bo')
plt.plot(days, yearly_temps(days, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```

##### Importing required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

##### Importing Titanic Data 

```python
URL = "https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
titanic = pd.read_csv(URL)
titanic.head()
```

**Task 1** : Create a pie chart presenting the male/female proportion


```python
titanic["sex"].value_counts().plot.pie(autopct="%.2f")
```

**Taks 2** : Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
fare_age_female = titanic[titanic['sex']=='female'][['fare', 'age']]
fare_age_male = titanic[titanic['sex']=='male'][['fare', 'age']]
```

```python
plt.scatter(x=fare_age_female['fare'], y=fare_age_female['age'], c='red')
plt.scatter(x=fare_age_male['fare'], y=fare_age_male['age'], c='blue')
plt.show()
```
